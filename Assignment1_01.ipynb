**TASK1: Using requests and BeautifulSoup to scrape a static html webpdage http://yuhenghu.com/ids566.html

```python
#pip install requests
import requests 

#pip install beautifulsoap4
from bs4 import BeautifulSoup as bs 
```

```python
#Load the Web page content
r = requests.get("http://yuhenghu.com/ids566.html")

#convert to beautiful Soup obj
soup = bs(r.content)

#print out our html
print(soup)

#html req in more readable form
print(soup.prettify())
```

```python
first_header = soup.find_all("td")
first_header
```

By using **find_all** extracting: Weekly Schedule from webpage 

```python
for tr in soup.find_all('tr'):
    values = [data for data in tr.find_all('td')]
    for value in values:
        print(value.text, '\t')
```
